# Load and save a subset of NOAA Optimum Interpolation v2.1 SST

This is the recipe for loading and downloading satellite SST data derived from the NOAA OISST v2.1 dataset:  
https://www.ncei.noaa.gov/metadata/geoportal/rest/metadata/item/gov.noaa.ncdc:C01606/html

Users can specify a time period, box coordinates and the filename they wish to create or update. Note that the product starts in September 1st 1981, which is the default value, and is regularly updated. By default, we set the end date to 60 days before the date of use of this notebook, as the product is updated with a certain delay.

To update an existing file, say after two years, the user needs to make sure that the `filename`  in the function is the same as the name of the file they want to update, and that the file is present in the intake-OISST directory.  
To print the current directory, the user can type:  
<pre> > import os <br>
 > os.getcwd()</pre>
 
The user also needs to make sure that the box coordinates passed to the function and the ones in the file they want to update are the same.

The data can be saved as a netcdf file using the subscript `filename.nc` or as a csv using `filename.csv`. Remember that saving large datasets as csv might not be practicle. 

```python
import urllib
import xarray as xr
import io
import datetime as dt
import numpy as np
import os
import os.path

os.chdir('intake-OISST')

from utils import download_NOAA_OISST

#today's date
now = dt.datetime.now()

# change argument in the function
ds = download_NOAA_OISST(lon_window = [100,116],lat_window = [-32,-20],start=dt.datetime(1981,9,1,12,0,0),end = now,filename = "test.nc")
```
